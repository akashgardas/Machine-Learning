## Subplots

```python
nums = np.arange(1, 6)
squares = np.power(nums, 2)
cubes = np.power(nums, 3)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1) # row 1, cols 2, 1st plot
plt.plot(nums, squares, 'ro--') #
plt.title('Squares')
plt.xlabel('Number')
plt.ylabel('Square')

plt.subplot(1, 2, 2) # row 1, cols 2, 2nd plot
plt.plot(nums, cubes, 'g^')
plt.title('Cubes')
plt.xlabel('Number')
plt.ylabel('Cube')

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y1 = np.sin(x**2)
y2 = np.cos(x**2)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'b')
plt.title('Sine Wave')

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r')
plt.title('Cosine Wave')

plt.tight_layout()
plt.show()
```

## Data Preprocessing

### Loading Dataset

```python
df = pd.read_csv('drive/MyDrive/Tekworks/Pulse_Calories.csv')
df.head()
```

### Handling Missing values

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Calories'] = df['Calories'].fillna(value=df['Calories'].mean())
df.describe()
```

```python
df['Date'] = df['Date'].fillna(method='ffill')
df.isna().sum()
```

### Datatype conversion

```python
df.info()
```

```python
# datatype conversion
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

### Handling Duplicates

```python
# removing duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

### Removing Outliers

```python
# removing outliers - Duration
df['Duration'] = df['Duration'].clip(lower=df['Duration'].quantile(0.05), upper=df['Duration'].quantile(0.95))
df.describe()
```

### Sorting

```python
# sorting by date
df.sort_values(by='Date', inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Pie chart
- Create a pie chart showing the percentage breakdown of Maxpulse into ranges (e.g., <=120, 121â€“140, 141â€“160, >160) and save as pie_maxpulse_ranges.png.

```python
df['Maxpulse'].describe()
```

```python
# defining ranges
bins=[0, 120, 140, 160, df['Maxpulse'].max()]
labels = ['<=120', '121-140', '141-160', '>160']

# categorizing data into ranges
df['Maxpulse_Range'] = pd.cut(df['Maxpulse'], bins=bins, labels=labels, right=True)
df.head()
```

```python
# value counts
maxpulse_counts = df['Maxpulse_Range'].value_counts()
maxpulse_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(maxpulse_counts, labels=maxpulse_counts.index, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title('Maxpulse Breakdown by Range')
# plt.savefig('pie_maxpulse_ranges.png')
plt.show()
```
