```python
import seaborn as sns
import matplotlib.pyplot as plt # not mandatory
import pandas as pd
```

# Loading Datasets

```python
# tips dataset
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
# Iris dataset
iris_df = sns.load_dataset('iris')
iris_df.head()
```

# Data Visualization

## Distribution Plots

### Histogram Plot

```python
sns.histplot(data=tips_df, x='total_bill')
plt.title('Distribution of Total Bill')
plt.show()
```

### Dis Plot

```python
sns.displot(data=tips_df, x='total_bill')
plt.title('Distribution of Total Bill')
plt.show()
```

### KDE Plot

```python
# KDE Plot
sns.kdeplot(data=tips_df, x='total_bill')
plt.title('KDE Plot - Smooth Distribution of Total Bill')
plt.show()
```

## 3. Categorical Plots
(Why: compare groups/categories)

- **Barplot** (`sns.barplot`): shows aggregate (mean) for categories.
- **Countplot** (`sns.countplot`): shows counts of categorical values.
- **Boxplot** (`sns.boxplot`): shows median, quartiles, and outliers.
- **Violinplot** (`sns.violinplot`): boxplot + KDE.

Use `hue` to split by another categorical variable.

### Bar Plot

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.barplot(data=tips_df, x='sex', y='total_bill')
plt.title('Average Total Bill by Sex')

plt.subplot(1, 2, 2)
sns.barplot(data=tips_df, x='sex', y='tip')
plt.title('Average Tip by Sex')

plt.tight_layout()
plt.show()
```

### Count Plot

```python
# Count Plot - Frequency of Days
sns.countplot(x='day', data=tips_df)
plt.title('Count Plot - Frequency of Days')
plt.show()
```

### Box Plot

```python
# Box Plot - Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=tips_df)
plt.title('Box Plot - Total Bill by Day')
plt.show()
```

### Violin Plot

```python
sns.violinplot(data=tips_df, x='day', y='total_bill')
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

## 4. Relationship Plots
(Why: see how two numeric variables relate)
- Scatterplot (sns.scatterplot): point cloud to inspect correlation.
- Lineplot (sns.lineplot): trends over an ordered variable.
- Relplot (sns.relplot): figure-level interface for relational plots; supports faceting.

### Scatter Plot

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.scatterplot(data=tips_df, x='total_bill', y='tip')  # without hue
plt.title('Scatter Plot - Total Bill vs Tip')

plt.subplot(1, 2, 2)
sns.scatterplot(data=tips_df, x='total_bill', y='tip', hue='sex') # with hue
plt.title('Scatter Plot - Total Bill vs Tip (Colored by Sex)')

plt.tight_layout()
plt.show()
```

### Line Plot

```python
sns.lineplot(data=tips_df, x='size', y='tip')
plt.title('Line Plot - Size vs Tip')
plt.show()
```

### Rel Plot

```python
# Relplot with Facets
sns.relplot(x='total_bill', y='tip', hue='sex', col='day', data=tips_df)
plt.suptitle('Relplot - Tip vs Total Bill by Day & Gender', y=1.02)
plt.show()
```

## Matrix & Correlation Plots
(Why: inspect relationships between many variables)
- Heatmap (sns.heatmap): visualize correlation matrix or any 2D numerical table.
- Clustermap (sns.clustermap): heatmap + hierarchical clustering to find groups.

```python
# Select only numeric columns
iris_num = iris_df.select_dtypes(include=['float', 'int'])

# Compute correlation
corr = iris_num.corr()
```

### Heat Map

```python
# Plot the heatmap
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation of Iris Features")
plt.show()
```

### Cluster Map

```python
# Select only numeric columns (exclude 'species')
iris_num = iris_df.select_dtypes(include=['float', 'int'])

# Compute correlation matrix
corr = iris_num.corr()

# Create the clustermap
sns.clustermap(corr, cmap='coolwarm', annot=True)
plt.suptitle('Clustermap - Grouped Correlations (Numeric Only)', y=1.02)
plt.show()
```
